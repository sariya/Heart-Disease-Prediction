```python
## Perform Logistic Regression
df_cont=df[['age','trestbps','chol','thalach','oldpeak']]

## perform standard scaling on cont data

## seperate the x(predictors) and y (target)

## perform the train and test split

## fit the train data on logistic regression

## Evaluate the model usung classification and ROC
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df.loc[:, ['age','trestbps','chol','thalach','oldpeak']]= scaler.fit_transform(df_cont)
```

```python
df.head()
```

```python
X= df.drop('target', axis = 1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size =0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(train_X, train_y)
```

```python
logreg.score(test_X, test_y)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y, logreg.predict(test_X)))
```

##### - if the model is predicting the person will not have an attack then the precision is 90% and if the model is predicting the person will
##### have an attach then the precision of the model is 84 % correct.
##### - recall is intepreted for class 1 as -if the person will actually have a heart diease it will be 92% accurate .
